```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Model performance metrics with comments
model_metrics = {
    "Model": ["Balanced Random Forest", "Decision Tree", "Logistic Regression", "Tuned Balanced RF"],
    "ROC AUC": [0.9750, 0.9421, 0.5403, 0.9521],
    "Recall": [0.91, 0.94, 0.00, 0.82],
    "Precision": [0.08, 0.09, 0.00, 0.06],
    "F1-Score": [0.15, 0.16, 0.00, 0.12],
    "Comments": [
        "Best model - High Recall & ROC AUC, but low precision", 
        "Decent model - High recall, but many false positives", 
        "Worst - Completely failed to detect fraud", 
        "Decent - But slightly worse than standard Balanced RF"
    ]
    
}

# Convert to DataFrame
df_metrics = pd.DataFrame(model_metrics)

# Save to CSV with comments and features included
metrics_file_path = "/Users/shivanginimarjiwe/Desktop/repo/DataScienceGuidedCapstone/Capstone2/model_metrics.csv"
df_metrics.to_csv(metrics_file_path, index=False)

# Visualization: Model Performance Comparison
plt.figure(figsize=(10, 6))
df_metrics.set_index("Model")[["ROC AUC", "Recall", "Precision", "F1-Score"]].plot(kind="bar", figsize=(10, 6))
plt.title("Model Performance Comparison")
plt.ylabel("Score")
plt.xticks(rotation=45)
plt.legend(title="Metrics")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Display DataFrame
print("\nModel Performance Metrics:")
print(df_metrics)

# Provide the CSV file path
print(f"\n Model Metrics file saved as: {metrics_file_path}")
```

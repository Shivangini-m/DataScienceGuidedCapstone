```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/Users/shivanginimarjiwe/Desktop/repo/DataScienceGuidedCapstone/Chem Corp Case Study/ChemCorp Sales Data.csv')


# Display basic info
df.info()
df.head()
```

```python
# Check for missing values in each column
df.isnull().sum()
```

```python
# State has missing values, decide  to fill them with "Unknown" instead of dropping those rows.
df['State'] = df['State'].fillna('Unknown')

```

```python
#Convert "Date" Column to Date Format

df['Date'] = pd.to_datetime(df['Date'])
```

```python
#Convert Financial Columns to Numeric

#The columns like "Cost (per unit)", "Price (per unit)", "Revenue (total)", "Cost (total)" have "$" signs that need to be removed.
# Remove $ symbols and convert to float


# Trim spaces from column names to prevent KeyErrors
df.columns = df.columns.str.strip()

# Define the correct column names
currency_columns = ['Cost (per unit)', 'Price (per unit)', 'Revenue (total)', 'Cost (total)']

# Replace non-numeric values ('-', empty spaces) with '0' before conversion
for col in currency_columns:
    df[col] = df[col].replace(r'[\$, -]', '', regex=True).replace('', '0').astype(float)

# Verify changes
df[currency_columns].info()
df.head()

```

```python
#Create Additional Columns (Optional)
# calculate Profit for better analysis in Tableau.

df['Profit'] = df['Revenue (total)'] - df['Cost (total)']
df['Profit Margin %'] = (df['Profit'] / df['Revenue (total)']) * 100
```

```python
#Save the Cleaned Dataset for Tableau

cleaned_file_path = "/Users/shivanginimarjiwe/Desktop/repo/DataScienceGuidedCapstone/Chem Corp Case Study/ChemCorp_Cleaned_Data.csv"
df.to_csv(cleaned_file_path, index=False)
print("Cleaned data saved successfully!")
df.head()
```

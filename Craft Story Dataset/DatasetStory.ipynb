```python
import pandas as pd
```

```python
# Load the dataset

df = pd.read_csv('/Users/shivanginimarjiwe/Desktop/repo/DataScienceGuidedCapstone/Craft Story Dataset/Bank Customer Churn Prediction.csv')

# Display basic info about the dataset
df.info(), df.head()
```

```python
'''
Initial Data Exploration

The dataset contains 10,000 bank customers with 12 features. Hereâ€™s a quick breakdown of what we have:

1. Features in the Dataset

Customer Demographics:
     customer_id (Unique ID)
     country (France, Spain, Germany)
     gender (Male, Female)
     age
Banking Information:
    credit_score
    tenure (Years with the bank)
    balance (Bank account balance)
    products_number (Number of bank products)
    credit_card (Has a credit card? 1 = Yes, 0 = No)
    active_member (Active membership status: 1 = Yes, 0 = No)
    estimated_salary
Target Variable:
churn (1 = Customer left, 0 = Stayed)

2. Key Observations
    No missing values, which is perfect for storytelling.
    The dataset includes both numerical and categorical variables, allowing diverse insights.
    The target variable (churn) is binary, so we can explore factors influencing customer retention.
    '''
```

```python

'''
Next Step:

Step 1: Analyze churn distribution (How many customers left vs. stayed).
Step 2: Identify trends & key drivers (e.g., Age, Credit Score, Balance, Active Status).
Step 3: Create visualizations to support the story
'''
```

```python
import matplotlib.pyplot as plt

# Churn distribution plot
plt.figure(figsize=(6, 4))
df['churn'].value_counts().plot(kind='bar', color=['blue', 'red'])
plt.xticks(ticks=[0, 1], labels=['Stayed (0)', 'Churned (1)'], rotation=0)
plt.xlabel("Customer Status")
plt.ylabel("Count")
plt.title("Churn Distribution in the Dataset")
plt.show()
```

```python
#The churn dataset is imbalancedâ€”more customers stayed with the bank than those who left.
```

```python
'''
Next Step

Step 4: Break down churn by key factors:
        Does churn correlate with age?
        Does credit score impact retention?
        Do active members churn less?
Step 5: Visualize trends using histograms and bar charts.
'''
```

```python
# Creating subplots to analyze churn by Age, Credit Score, and Active Membership
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Churn by Age
df[df['churn'] == 1]['age'].hist(bins=30, alpha=0.7, color='red', ax=axes[0])
df[df['churn'] == 0]['age'].hist(bins=30, alpha=0.7, color='blue', ax=axes[0])
axes[0].set_title("Churn by Age")
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Count")
axes[0].legend(["Churned", "Stayed"])

# Churn by Credit Score
df[df['churn'] == 1]['credit_score'].hist(bins=30, alpha=0.7, color='red', ax=axes[1])
df[df['churn'] == 0]['credit_score'].hist(bins=30, alpha=0.7, color='blue', ax=axes[1])
axes[1].set_title("Churn by Credit Score")
axes[1].set_xlabel("Credit Score")
axes[1].set_ylabel("Count")
axes[1].legend(["Churned", "Stayed"])

# Churn by Active Membership
df.groupby("active_member")["churn"].mean().plot(kind='bar', color=['blue', 'red'], ax=axes[2])
axes[2].set_title("Churn Rate by Active Membership")
axes[2].set_xlabel("Active Member (0 = No, 1 = Yes)")
axes[2].set_ylabel("Churn Rate")

plt.tight_layout()
plt.show()
```

```python
'''
Key Insights from Churn Analysis

1. Churn by Age:
   Older customers churn more frequently, especially after age 40.
   Younger customers seem more stable, with fewer leaving.

2. Churn by Credit Score:
   Customers with lower credit scores are more likely to churn.
   High-credit-score customers tend to stay.

3. Churn by Active Membership:
   Active members churn significantly less than inactive members.
   This suggests that engagement with the bank reduces churn.
'''
```

```python
'''
Next Steps

Step 6: Breakdown churn by country (Are some regions more likely to churn?)
Step 7: Analyze account balance vs. churn (Do customers with higher balances stay longer?)
Step 8: Create a PowerPoint or Jupyter Notebook summarizing findings
'''
```

```python


import matplotlib.pyplot as plt
import seaborn as sns

# Churn rate by country
country_churn = df.groupby("country")["churn"].mean().sort_values()

# Plot churn rate by country
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="country", y="churn", estimator=lambda x: sum(x)/len(x), palette="coolwarm")

plt.title("Churn Rate by Country")
plt.xlabel("Country")
plt.ylabel("Churn Rate")
plt.show()

# Churn vs. Account Balance Distribution
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="churn", y="balance", palette="coolwarm")

plt.xticks(ticks=[0, 1], labels=['Stayed (0)', 'Churned (1)'])
plt.title("Distribution of Account Balance by Churn Status")
plt.xlabel("Churn Status")
plt.ylabel("Account Balance")
plt.show()

```

```python
'''
Key Insights from Analysis

1. Churn by Country: 
   Germany has the highest churn rate compared to France and Spain.
   Customers in France and Spain are more stable with lower churn rates.
   The bank may need targeted retention strategies for German customers.

2. Churn vs. Account Balance:
   Median account balance is similar for both churned and retained customers.
   Some high-balance customers still churn, suggesting balance alone is not a strong predictor.
   Additional factors (e.g., engagement, products used) might be influencing churn.
'''   
```

```python
# Generate a scatterplot for Credit Score vs. Balance with Churn as hue
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="credit_score", y="balance", hue="churn", alpha=0.6, palette="coolwarm")

plt.title("Credit Score vs. Balance (Colored by Churn)")
plt.xlabel("Credit Score")
plt.ylabel("Account Balance")
plt.legend(title="Churn", labels=["Stayed (0)", "Churned (1)"])
plt.show()

# Generate a correlation heatmap ( Drop non-numeric columns)
plt.figure(figsize=(10, 6))

# Convert categorical variables to numeric or remove them
numeric_df = df.select_dtypes(include=['number'])  # Select only numeric columns

# Compute correlation
correlation_matrix = numeric_df.corr()

# Generate heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap of Features")
plt.show()
```

```python
'''
The correlation heatmap visually represents the relationships between different numerical features in the dataset. 

The color scale on the right shows the correlation values:

Red (near 1.0): Strong positive correlation.
Blue (near -1.0): Strong negative correlation.
Near 0: No strong relationship between the variables

Insights from Additional Analysis

1. Scatterplot: Credit Score vs. Balance (Colored by Churn)
     The scatterplot shows that churned customers (blue) are spread across all credit scores and balance levels,
     meaning that balance alone is not a strong predictor of churn.
     However, there is a concentration of churned customers in the lower credit score range.

2. Correlation Heatmap
     Age has the highest correlation with churn (0.29)  â€”   older customers are more likely to leave.
     Balance has a weak correlation with churn (0.12), meaning high balance does not guarantee retention.
     Active Membership (-0.16) is negatively correlated with churn, meaning active members are less likely to leave.
     Credit Score (-0.03) has almost no correlation with churn, which is surprising!
'''
```
